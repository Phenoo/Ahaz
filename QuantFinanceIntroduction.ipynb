# Mathematical Tools


```python
import numpy as np
import pandas as pd
from pylab import mpl, plt

mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
def create_plot(x, y, styles, labels, axlabels):
    plt.figure(figsize=(10, 6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
        plt.legend(loc=0)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
create_plot([x], [f(x)], ['b'], ['f(x)'], ['x', 'f(x)'])
```

```python
res = np.polyfit(x, f(x), deg=1, full=True)
```

```python
res
```

```python
ry = np.polyval(res[0], x)
```

```python
create_plot([x, x], [f(x), ry], ['b', 'r.'], ['f(x)', 'regression'], ['x', 'f(x)'])
```

```python
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)
```

```python
create_plot([x, x], [f(x), ry], ['b', 'r.'], ['f(x)', 'regression'], ['x', 'f(x)'])
```

```python
reg = np.polyfit(x, f(x), 7)
ry =np.polyval(reg, x)
```

```python
np.allclose(f(x), ry)
```

```python
np.mean((f(x) - ry) ** 2)
```

```python
create_plot([x, x], [f(x), ry], ['b', 'r.'], ['f(x)', 'regression'], ['x', 'f(x)'])
```

```python
matrix = np.zeros((3 + 1, len(x)))
matrix[3, :] = x ** 3
matrix[2, :] = x ** 2
matrix[1, :] = x
matrix[0, :] = 1
```

```python
reg = np.linalg.lstsq(matrix.T, f(x), rcond=None)[0]
```

```python
reg.round(4)
```

```python
ry = np.dot(reg, matrix)
```

```python
create_plot([x, x], [f(x), ry], ['b', 'r.'], ['f(x)', 'regression'], ['x', 'f(x)'])
```

```python
matrix[3, :] = np.sin(x)
```

```python
reg = np.linalg.lstsq(matrix.T, f(x), rcond=None)[0]
```

```python
reg.round(4)
```

```python
ry = np.dot(reg, matrix)
```

```python
def fm(p):
  x, y = p
  return np.sin(x) + 0.25 * x + np.sqrt(y) + 0.005 * y ** 2
```

```python
x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x, y)
```

```python
Z = fm((X, Y))
x = X.flatten()
y = Y.flatten()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap='coolwarm', linewidth=0.5, antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
matrix = np.zeros((len(x), 6 + 1))
matrix[:, 6] = np.sqrt(y)
matrix[:, 5] = np.sin(x)
matrix[:, 4] = y ** 2
matrix[:, 3] = x ** 2
matrix[:, 2] = y
matrix[:, 1] = x
matrix[:, 0] = 1
```

```python
reg = np.linalg.lstsq(matrix, fm((x, y)), rcond=None)[0]
```

```python
RZ = np.dot(matrix, reg).reshape((20, 20))
```

```python
from mpl_toolkits.mplot3d.axes3d import proj3d
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf1 = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap=mpl.cm.coolwarm, linewidth=0.5, antialiased=True)
surf2 = ax.plot_surface(X, Y, RZ, rstride=2, cstride=2, label='regression')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.legend()
fig.colorbar(surf, shrink=0.5, aspect=5)
```

# Convex Optimization

```python
def fm(p):
  x, y = p
  return (np.sin(x) + 0.5 * x ** 2) + (np.sin(y) + 0.5 * y ** 2)
```

```python
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)
Z = fm((X, Y))
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap='coolwarm', linewidth=0.5, antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

# Constrained Optimization

```python
import math
```

```python
def EU(p):
  s, b = p
  return -(0.5 * math.sqrt(s * 15 * b * 5) + 0.5 * math.sqrt(s * 5 * b * 12))
```

```python
cons = ({'type': 'ineq', 'fun': lambda p: 100 - p[0] * 10 - p[1] * 10})
```

```python
bnds = ((0, 1000), (0, 1000))
```

```python
result = sco.minimize(EU, [5, 5], method='SLSQP', bounds=bnds, constraints=cons)
```

```python
result
```

```python
result['fun']
```

# Integration

```python
import scipy.integrate as sci
```

```python
def f(x):
  return np.sin(x) + 0.5 * x
```

```python
x = np.linspace(0, 10)
y = f(x)
a = 0.5
b = 9.5
Ix = np.linspace(a, b)
Iy = f(Ix)
```

```python
from matplotlib.patches import Polygon
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y,'b', linewidth=2)
plt.ylim(bottom=0)
Ix = np.linspace(a, b)
Iy = f(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.75 * (a + b), 1.5, r"$\int_a^b f(x)dx$",
horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075,'$x$')
plt.figtext(0.075, 0.9,'$f(x)$')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$','$b$'))
ax.set_yticks([f(a), f(b)]);
```
